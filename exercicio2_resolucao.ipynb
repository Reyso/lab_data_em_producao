## 1 - Crie uma nova coluna chamda: "house_age"

    - se o valor 'date' > 2014-01-01, 'new_house
    - se o valor 'date' < 2014-01-01, 'old_house

```python
data['house_age'] = np.where(data['date'] > '2014-01-01' , 'new_house','old_house')
#data['house_age'] = data['date'].apply( lambda x: 'new_house' if x > pd.to_datetime('2014-01-01',format='%y-%m-%d') else 'old_house' )
```

## 2 - Crie uma nova coluna chamda: "dormitory_type"

    -se bedrooms == 1 , 'studio'
    -se badrooms == 2 , 'apartament'
    -se badrroms > 2 , 'house'

```python
data['bedrooms'].unique() #temos apartamentos sem quartos
```

```python
data['dormitory_type'] = data['bedrooms'].apply( lambda x: 'studio' if x ==1 else
                                                           'apartament' if x ==2 else
                                                           'house' if x >2 else 'NAN')
```

```python
data.head()
```

## 3 -  Crie uma nova coluna chamda: "condition_type"

    -se condition <= 2 , bad
    -se condidtion ==3 ou condidtion ==4 , 'regular
    -se condidtion == 5 , good

```python
data['condition_type'] = data['condition'].apply( lambda x: 'bad' if x <=2 else
                                                            'regular' if (x == 3) or (x == 4) else
                                                            'good' if x == 5 else 'NAN')
```
