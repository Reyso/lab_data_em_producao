```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

plt.style.use('fivethirtyeight')
```

```python
data = pd.read_csv('dataset/kc_house_data.csv')
```

## 1- Crie uma nova coluna chamada: 'dormitory_type'
    - se o valor da coluna "bedrooms" == 1 -> 'studio'
    - se o valor da coluna "bedrooms" == 2 -> 'apartament'
    - se o valor da coluna "bedrooms" > 2 -> 'house'

```python
data['dormitory_type'] = data['bedrooms'].apply([lambda x : 'studio' if x ==1 else
                                                                        'apartament' if x == 2 else
                                                                        'house' if x >2 else
                                                                         np.NaN])
```

```python
data
```

## 2 - faça um gŕafico de barras que repesente a soma dos preços pelos números de quartos

```python
data_temp = data[['bedrooms','price']].groupby('bedrooms').sum().reset_index(); data_temp
```

```python
sns.barplot(data = data_temp , 
            x = 'bedrooms',
            y = 'price')
```

## 3 - Faça um gráfico de linhas que represente a média dos preços pelo ano de construção dos imóveis

```python
data_temp = data[['price','yr_built']].groupby('yr_built').mean().reset_index()
```

```python
data_temp['price'].max()
```

```python
plt.figure(figsize=(10,5))
fig1 = sns.lineplot(data = data_temp,
            x = 'yr_built',
            y = 'price',
            legend='auto')
xticks = np.arange(300000,782050.0,20000)
yticks = np.arange(1900,2021,10)
plt.yticks(xticks)
plt.xticks(yticks,rotation = 0)
plt.show()
```

## 4 - faça um gráfico de barras que represente a média dos preços pelo tipo dos dormitório


```python
data_temp = data[['price','dormitory_type']].groupby('dormitory_type').mean().reset_index(); data_temp
```

```python
fig3 = plt.figure(figsize=(10,5))

sns.barplot(data = data_temp,
           x = 'dormitory_type',
           y = 'price')
plt.yticks(np.arange(50000,603092,60000))
plt.title('Média dos preços por dormitório')
```
